```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('final_train2.csv')
test = pd.read_csv('final_test2.csv')
df = pd.concat([train, test], axis = 0)
```

```python
train.shape
```

```python
def p25(series):
    return series.quantile(0.25)

def p75(series):
    return series.quantile(0.75)
```

```python
book = {}
for brand in df['oem'].unique():
    var_name = f"{brand}"
    temp = df.loc[df.oem == brand]
    count = temp['model'].value_counts()
    count.columns = ['model', 'count']
    ana = temp.groupby(['model'])['listed_price'].agg([np.mean, min, p25, p75, max])
    book[var_name] = pd.concat([ana, count], axis = 1) 
```

```python
indices = []
for model in df['model'].unique():
        temp2 = df.loc[df.model == model]
        #Creating an upper bound to drop outliers
#A good assumption for high outliers would be 
# greater than (0.75 percentile) + 0.25 *(0.75 percentile) 
        _75p = temp2['listed_price'].quantile(0.75)
        upper_bound = 1.25 * _75p
        for i in temp2.index:
            if (temp2['listed_price'][i] >= upper_bound).all():
                indices.append(i)
```

```python
max(indices)#therefore all indices are in train itself
train = train.drop(indices, axis = 0)
```
